```python
eligible_venue = df['venue'].value_counts()[df['venue'].value_counts() > 480].index.tolist()
```

```python
df = df[df['venue'].isin(eligible_venue)]
```

```python
df.shape
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df.head()
```

```python
df.tail()
```

```python
df['ball_no'] = df['ball'].apply(lambda x:str(x).split('.')[1])
df['over'] = df['ball'].apply(lambda x:str(x).split('.')[0])
df['ball_bowled'] = (df['over'].astype(int)*6 + df['ball_no'].astype(int))
df['ball_left'] = 120 - df['ball_bowled']
```

```python
df['ball_left'] = df['ball_left'].apply(lambda x:0 if x<0 else x)
```

```python
df.tail()
```

```python
df.info()
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:1 if x!='0' else '0')
```

```python
df.head()
```

```python
df.shape
```

```python
df['player_dismissed'] = df['player_dismissed'].astype(int)
df['cumulative_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
df['wicket_left'] = 10 - df['player_dismissed']
```

```python
df.head()
```

```python
df['current_run_rate'] = (df['current_score']*6)/df['ball_bowled']
```

```python
df.tail()
```

```python
# group by 'match id'
groups = df.groupby('match_id')

# 5 over = 30 balls
match_id = df['match_id'].unique()
last_five = []

for id in match_id:
    # Apply rolling sum on the 'runs' column
    rolling_sums = groups.get_group(id)['runs'].rolling(window=30).sum()
    last_five.extend(rolling_sums.values.tolist())
```

```python
df['last_five'] = last_five
```

```python
df['last_five']
```

```python
df.head()
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df.head()
```

```python
df['batting_team'].nunique()
```

```python
final_df.columns
```

```python
final_df = final_df[['batting_team', 'bowling_team', 'venue', 'current_score', 'ball_left', 'wicket_left', 'current_run_rate', 'last_five', 'runs_x']]
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df.shape
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
x = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
# import machine learnign algorithm
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBRegressor
```

```python
lr = LinearRegression()
dt = DecisionTreeClassifier(max_depth=5)
rf = RandomForestClassifier(n_estimators=50,random_state=2,max_depth=5)
xr = XGBRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12, random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
```

```python
transformer = ColumnTransformer([
    ('transformer', OneHotEncoder(drop='first', sparse_output=False), ['batting_team', 'bowling_team', 'venue'])
], remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1', transformer),
    ('step2', StandardScaler(with_mean=False)),
    ('step3', XGBRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12, random_state=1))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```
